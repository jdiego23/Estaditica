```python
import numpy as np
import os
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import tarfile
import zipfile
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import AgglomerativeClustering
from scipy.io import loadmat
from urllib.request import urlretrieve
from os.path import isfile, isdir
import plotly.figure_factory as ff
import plotly.express as px
import statsmodels.api as sm
from IPython.display import Image
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.utils import to_categorical
import glob
import random
import matplotlib.image as img
from PIL import Image
#Install pillow
from PIL import Image
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
%matplotlib inline
```

```python
CarpetaEntrenamiento = 'data/train'
categorias = ['Black-grass', 'Charlock', 'Cleavers', 'Common Chickweed', 'Common wheat', 'Fat Hen',
              'Loose Silky-bent', 'Maize', 'Scentless Mayweed', 'Shepherds Purse',
              'Small-flowered Cranesbill', 'Sugar beet']
```

```python
imagenes = []
etiquetas = []
```

```python
for categoria in categorias:
    carpeta = os.path.join(CarpetaEntrenamiento, categoria)
    # Se itera sobre los archivos de imagen en cada carpeta
    for nombreArchivo in os.listdir(carpeta):
        # Se lee la imagen y convertirla a escala de grises
        rutaImagen = os.path.join(carpeta, nombreArchivo)
        imagen = Image.open(rutaImagen).convert('L')
        # Se redimensiona la imagen a un tamao fijo, por ejemplo (128, 128)
        imagen = imagen.resize((128, 128))
        # Se convierte la imagen a un array numpy
        vectoImagenes = np.array(imagen)
        # Se agrega la imagen y la etiqueta a las listas
        imagenes.append(vectoImagenes)
        etiquetas.append(categoria)
```

```python
imagenes = np.array(imagenes)
etiquetas = np.array(etiquetas)
```

```python
etiquetaEncoder = LabelEncoder()
etiquetasEncoder = etiquetaEncoder.fit_transform(etiquetas)
```

```python
etiquetasCategorical = to_categorical(etiquetasEncoder)
```

```python
X_train, X_test, y_train, y_test = train_test_split(imagenes, etiquetasCategorical, test_size=0.2, random_state=42)
```

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
modelo = Sequential()
modelo.add(Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 1)))
modelo.add(MaxPooling2D((2, 2)))
modelo.add(Conv2D(64, (3, 3), activation='relu'))
modelo.add(MaxPooling2D((2, 2)))
modelo.add(Flatten())
modelo.add(Dense(64, activation='relu'))
modelo.add(Dense(len(categorias), activation='softmax'))
```

```python
modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
modelo.fit(X_train, y_train, epochs=20, batch_size=30, validation_data=(X_test, y_test))
```

```python
carpetaPruebas = 'data/test'
imagenesPruebas = []
```

```python
for nobmbreCarpeta in os.listdir(carpetaPruebas):
    rutaImagen = os.path.join(carpetaPruebas, nobmbreCarpeta)
    imagen = Image.open(rutaImagen).convert('L')
    imagen = imagen.resize((128, 128))
    vectorImagen = np.array(imagen)
    imagenesPruebas.append(vectorImagen)
```

```python
imagenesPruebas = np.array(imagenesPruebas) / 255.0
```

```python
predicciones = modelo.predict(imagenesPruebas)
```

```python
etiquetasPredecidas = etiquetaEncoder.inverse_transform(np.argmax(predicciones, axis=1))
```

```python
resultados = pd.DataFrame({'file': os.listdir(carpetaPruebas), 'species': etiquetasPredecidas})
```

```python
archivoPredicion = os.path.join('data', 'predicciones.csv')
resultados.to_csv(archivoPredicion, index=False)
```
